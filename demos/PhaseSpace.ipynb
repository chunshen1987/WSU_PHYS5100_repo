```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

# Phase space diagram

## nodes, repeller, and saddle points

### 1. Real eigenvalues

```python
ll = np.linspace(-1.0, 1.0, 20)
x1, x2 = np.meshgrid(ll, ll)

v1 = 1.*x1
v2 = 2.*x2
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot([0.], [0,], marker='o', markersize=20)
ax.streamplot(x1, x2, v1, v2, density = 2, linewidth=1.5)
ax.set_xlabel(r"$x_1$", fontsize=20)
ax.set_ylabel(r"$x_2$", fontsize=20)
ax.set_title("repeller", fontsize=20)

v1 = -1.*x1
v2 = x1 - 2.*x2
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot([0.], [0,], marker='o', markersize=20)
ax.streamplot(x1, x2, v1, v2, density = 2, linewidth=1.5)
ax.set_xlabel(r"$x_1$", fontsize=20)
ax.set_ylabel(r"$x_2$", fontsize=20)
ax.set_title("node", fontsize=20);
```

```python
v1 = -1.*x1 + x2
v2 = x1 + 2.*x2
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot([0.], [0,], marker='o', markersize=20)
ax.streamplot(x1, x2, v1, v2, density = 2, linewidth=1.5)
ax.set_xlabel(r"$x_1$", fontsize=20)
ax.set_ylabel(r"$x_2$", fontsize=20)
ax.set_title("saddle", fontsize=20);
```

### 2. Complex eigenvalues

```python
v1 = x2
v2 = -x1
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot([0.], [0,], marker='o', markersize=20)
ax.streamplot(x1, x2, v1, v2, density = 2, linewidth=1.5)
ax.set_xlabel(r"$x_1$", fontsize=20)
ax.set_ylabel(r"$x_2$", fontsize=20)
ax.set_title("node", fontsize=20);

v1 = - x1 - x2
v2 = 2.*x1
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot([0.], [0,], marker='o', markersize=20)
ax.streamplot(x1, x2, v1, v2, density = 2, linewidth=1.5)
ax.set_xlabel(r"$x_1$", fontsize=20)
ax.set_ylabel(r"$x_2$", fontsize=20)
ax.set_title("Spiral node", fontsize=20);

v1 = x1 - x2
v2 = 2.*x1
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot([0.], [0,], marker='o', markersize=20)
ax.streamplot(x1, x2, v1, v2, density = 2, linewidth=1.5)
ax.set_xlabel(r"$x_1$", fontsize=20)
ax.set_ylabel(r"$x_2$", fontsize=20)
ax.set_title("Spiral repeller", fontsize=20);
```

### 3. Degenerate eigenvalue

```python
v1 = 2.*x1 + x2
v2 = 2.*x2
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot([0.], [0,], marker='o', markersize=20)
ax.streamplot(x1, x2, v1, v2, density = 1.5, linewidth=1.5)
ax.set_xlabel(r"$x_1$", fontsize=20)
ax.set_ylabel(r"$x_2$", fontsize=20)
ax.set_title("Repeller", fontsize=20);

v1 = -2.*x1 + x2
v2 = -2.*x2
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot([0.], [0,], marker='o', markersize=20)
ax.streamplot(x1, x2, v1, v2, density = 1.5, linewidth=1.5)
ax.set_xlabel(r"$x_1$", fontsize=20)
ax.set_ylabel(r"$x_2$", fontsize=20)
ax.set_title("node", fontsize=20);
```

## Rabits and sheep

$$\dot{x} = x (3 - x - 2y)$$

$$\dot{y} = y (2 - y - x)$$

```python
ll = np.linspace(-1.0, 5.0, 61)
x, y = np.meshgrid(ll, ll)

xdot = x*(3 - x - 2.*y)
ydot = y*(2 - y - x)

fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter([0., 0, 3, 1], [0, 2, 0., 1], s=300, marker='o', color='r')
ax.streamplot(x, y, xdot, ydot, density = 1.5, linewidth=1.5)
ax.set_xlabel(r"Rabbits", fontsize=20)
ax.set_ylabel(r"Sheep", fontsize=20)
ax.set_title("Rabbits vs. sheep", fontsize=20);
```

## Pendulum

```python
theta = np.linspace(-3.5*np.pi, 3.5*np.pi, 100)
omega = np.linspace(-3, 3, 100)
theta, omega = np.meshgrid(theta, omega)

mu = 0.1
goverL = 1.

thetadot = omega
omegadot = -mu*omega - goverL*np.sin(theta)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(np.linspace(-3, 3, 7)*np.pi, np.zeros(7), s=300, c='r', )
ax.streamplot(theta, omega, thetadot, omegadot, density = 4, linewidth=1.5)
ax.set_xlabel(r"$\theta$", fontsize=20)
ax.set_ylabel(r"$\dot{\theta}$", fontsize=20)
ax.set_title("Pendulum", fontsize=20);
```
