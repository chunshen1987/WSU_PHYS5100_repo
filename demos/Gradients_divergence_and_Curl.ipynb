```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

# Demonstration of the gradients of a scalar field

## The field of gravitational potential and the gravitational force as its gradient

```python
ll = np.linspace(-1.0, 1.0, 20)
dx = ll[1] - ll[0]
x, y = np.meshgrid(ll, ll)
z = -1./np.sqrt(x**2 + y**2 + 0.1)

v, u = np.gradient(z, dx, dx)

fig, ax = plt.subplots(figsize=(7, 6))
c = ax.contourf(x, y, -z, 50, cmap="Reds")
plt.colorbar(c)
ax.quiver(x, y, -u, -v, width=0.005);
```

```python
grid = np.linspace(-2, 2, 31)
dx = grid[1] - grid[0]
x, y = np.meshgrid(grid, grid)

z = x * np.exp(-x**2 - y**2)

v, u = np.gradient(z, dx, dx)

fig, ax = plt.subplots(figsize=(9, 6))
c = ax.contourf(x, y, z, 50, cmap="jet")
plt.colorbar(c)
ax.quiver(x, y, u, v);
```

```python
grid = np.linspace(-2, 2, 31)
dx = grid[1] - grid[0]
x, y = np.meshgrid(grid, grid)

z = x * np.exp(-x**2 - y**2)

v, u = np.gradient(z, dx, dx)
w = np.sqrt(u**2. + v**2.)

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
c = ax.plot_surface(x, y, z, cmap="jet")
nskip = 2
ax.quiver(x[::nskip, ::nskip], y[::nskip, ::nskip], z[::nskip, ::nskip],
          u[::nskip, ::nskip], v[::nskip, ::nskip], w[::nskip, ::nskip],
          length=0.3, color='k');
```

# Demonstration of the divergence of a vector field

```python
ll = np.linspace(-2.0, 2.0, 21)
dx = ll[1] - ll[0]
x, y = np.meshgrid(ll, ll)

alpha = -2.0
u = alpha*x
v = alpha*y

uy, ux = np.gradient(u, dx, dx)
vy, vx = np.gradient(v, dx, dx)

levels = np.linspace(-8., 8, 51)

fig, ax = plt.subplots(figsize=(6, 6))
ax.contourf(x, y, ux + vy, levels, cmap="bwr")
ax.quiver(x, y, u, v, scale=70, width=0.005)
```

```python
grid = np.linspace(-2, 2, 31)
dx = grid[1] - grid[0]
x, y = np.meshgrid(grid, grid)

z = x * np.exp(-x**2 - y**2)

v, u = np.gradient(z, dx, dx)

uy, ux = np.gradient(u, dx, dx)
vy, vx = np.gradient(v, dx, dx)

fig, ax = plt.subplots(figsize=(9, 6))
c = ax.contourf(x, y, ux + vy, 50, cmap="jet")
plt.colorbar(c)
ax.quiver(x, y, u, v)
```

# Demonstration of the Curl of a vector field

## A rotational velocity field

```python
ll = np.linspace(-2.0, 2.0, 21)
dx = ll[1] - ll[0]
x, y = np.meshgrid(ll, ll)

omega = 1.0

u = -omega*y
v = omega*x
Z = omega*np.ones(u.shape)

levels = np.linspace(-4., 4, 51)
fig, ax = plt.subplots(figsize=(6, 6))
ax.contourf(x, y, Z, levels, cmap="bwr")
ax.quiver(x, y, u, v, color='k', scale = 20, width=0.003)
plt.xlim([-2, 2])
plt.ylim([-2, 2])
```

## A shear velocity field

```python
ll = np.linspace(-2.0, 2.0, 21)
dx = ll[1] - ll[0]
x, y = np.meshgrid(ll, ll)

alpha = 1.0
beta = 2.0

u = alpha*y
v = beta*x

Z = (beta - alpha)*np.ones(u.shape)

levels = np.linspace(-4., 4, 51)
fig, ax = plt.subplots(figsize=(6, 6))
ax.contourf(x, y, Z, levels, cmap="bwr")
ax.quiver(x, y, u, v, color='k', scale = 30, width=0.003)
plt.xlim([-2, 2])
plt.ylim([-2, 2])
```
